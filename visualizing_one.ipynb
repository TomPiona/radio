```python
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
import plotly.figure_factory as ff
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.DataFrame(columns=['artist', 'time', 'title', 'day', 'month', 'station', 'start_time', 'end_time'])

for file in ['94.9_900_songs.csv', 'second_94.9.csv', 'kmel_songs.csv', 'star_songs.csv', 'not_even_in_our_timezone.csv']:
    df = pd.read_csv(file)
    df = df[::-1].reset_index(drop=True)
    df['start_time'] = pd.to_datetime('2017' + (df.month*100+df.day).apply(str) + ' ' + df.time,format='%Y%m%d %I:%M %p')
    dt_2 = df['start_time'][1:].reset_index(drop=True)
    dt_2.set_value(value=df['start_time'].iloc[-1] + pd.Timedelta(minutes=3), label=len(dt_2))
    df['end_time'] = dt_2
    data = pd.concat([data, df])

print('Rows:', len(data))
data.head()
```

```python
plays_order = data.groupby('title').count().sort_values('artist', ascending=False).index
rp = data.groupby(['title', 'station']).count().reset_index()
rp['sm'] = pd.Categorical(
    rp['title'], 
    categories=plays_order, 
    ordered=True
)
rp.set_index(['title', 'station']).sort_values('sm')
```

```python
cpy = data.copy()
cpy = cpy[cpy['title'] == 'rockstar [Feat. 21 Savage]'].reset_index(drop=True)
cpy = cpy[['station', 'start_time', 'end_time']]
cpy.columns = ['Task', 'Start', 'Finish']
cpy.head()

ob = ff.create_gantt(cpy, group_tasks=True)
py.iplot(ob)
```
