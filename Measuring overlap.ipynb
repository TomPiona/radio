```python
import pandas as pd
```

```python
data = pd.DataFrame(columns=['artist', 'time', 'title', 'day', 'month', 'station', 'start_time', 'end_time'])

for file in list(map(lambda x: 'data/' + x, ['94.9_900_songs.csv', 'second_94.9.csv', 'kmel_songs.csv', 'star_songs.csv', 'not_even_in_our_timezone.csv'])):
    df = pd.read_csv(file)
    df = df[::-1].reset_index(drop=True)
    df['start_time'] = pd.to_datetime('2017' + (df.month*100+df.day).apply(str) + ' ' + df.time,format='%Y%m%d %I:%M %p')
    dt_2 = df['start_time'][1:].reset_index(drop=True)
    dt_2.set_value(value=df['start_time'].iloc[-1] + pd.Timedelta(minutes=3), label=len(dt_2))
    df['end_time'] = dt_2
    data = pd.concat([data, df])

print('Rows:', len(data))
data.head()
```

```python
def get_station(st):
    return data[data['station'] == st]
def get_song(df, sn):
    return df[df['title'] == sn]
def get_times(df):
    return df.sort_values('start_time')[['start_time', 'end_time']].values
```

```python
def overlap_counter(s1, s2):
    def f(ar1, ar2, overlaps):
        if len(ar1) == 0 or len(ar2) == 0:
            return overlaps
        x, y = ar1[0], ar2[0]
        print(x, '\n', y)
        print()
        print()
        if x[1] > y[0]:
            overlaps += 1
            if x[1] < y[1]:
                return f(ar2, ar1[1:], overlaps)
            else:
                if y[1] < ar1[0][0]:
                    return f(ar2, ar1[1:], overlaps)
                else:
                    return f(ar1, ar2[1:], overlaps)
        else:
            if len(ar1) > 1 and y[0] < ar1[1][0]:
                return f(ar2, ar1[1:], overlaps)
            else:
                return f(ar1[1:], ar2, overlaps)
    if s1[0][0] <= s2[0][0]:
        return f(s1, s2, 0)
    return f(s2, s1, 0)
```

```python
overlap_counter(get_times(get_song(get_station(94.9), 'rockstar [Feat. 21 Savage]')),
                get_times(get_song(get_station('kmel'), 'rockstar [Feat. 21 Savage]')))
```
